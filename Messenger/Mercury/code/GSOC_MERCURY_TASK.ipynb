```python
# taking the mean lat long of composition into a single var
mean_al = al_si.mean(axis=1)
mean_ca = ca_si.mean(axis=1)
mean_fe = fe_si.mean(axis=1)
mean_mg = mg_si.mean(axis=1)
mean_s = s_si.mean(axis=1)
top = top_mercury.mean(axis=1)
bottom = bottom_mercury.mean(axis=1)
```

```python
#create the custom dataframe with mean relation between the geographical datas
data = {'Al-Si ratio':mean_al,'Ca-Si ratio':mean_ca,'Fe-Si ratio':mean_fe,'Mg-Si ratio':mean_mg,'S-Si ratio':mean_s,'Top_mercury':top,'Bottom_mercury':bottom}
df = pd.DataFrame(data)
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators=100,criterion='mse')
```

```python
#creating dependent and independent variables.
x = df.drop('Bottom_mercury',axis=1)
y = df['Bottom_mercury']
```

```python
#import splitting library
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state=0)
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
#comparing prediction with true value.
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```

```python
#plotting the Histogram of test and predicted value

plt.subplot(1,2,1)
plt.hist(y_test,bins=10)
plt.title('Real Bottom  mercury value')
plt.subplot(1,2,2)
plt.hist(y_pred,bins=10)
plt.title('Predicted  Value')
plt.show()

```

```python
#plotting the resulting prediction and plot the residuals (differencefrom the true image) as a 2-D image
plt.figure(figsize=(8,8))
plt.scatter(y_test, y_pred, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(y_pred), max(y_test))
p2 = min(min(y_pred), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()

```
